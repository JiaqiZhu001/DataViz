For general public data viz,the content should be more generic and easy to under stand, so we should list top Apps based on its category and reviews, and intalls.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("./Data/googleplaystore.csv")
data
```

```python
data["Reviews"] = data["Reviews"].replace({"K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(int)
```

```python
data["Reviews"] = data["Reviews"].replace({"K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(int)
data["Installs"] = data["Installs"].str.replace(',', '')
data["Installs"] = data["Installs"].str.replace('+', '')
## remove only str value
data["Installs"] = data["Installs"].str.replace('Free', '')
data["Installs"] = pd.to_numeric(data["Installs"])
data = data[data['Rating'].notna()]
pd.to_numeric(data['Rating'])
data
```

Lets pick Top 20 apps with Rating and check their reviews and installs

```python
data["App"] = data["App"].str.strip()
data = data.drop_duplicates(subset=["App"],keep="first")
filter_reviews = data[data["Reviews"] > 10000000]
filter_reviews = filter_reviews.drop_duplicates(subset=['App'],keep="first")
top_rate = filter_reviews.nlargest(20,['Rating'])
top_rate
```

```python
ind = np.arange(len(top_rate))
width = 0.4

reviews = top_rate['Reviews']
installs = top_rate['Installs']
installs = installs.astype(int)
```

```python

fig, ax = plt.subplots()
ax.barh(ind, installs, width, color='red', label='Installs')
ax.barh(ind + width, reviews, width, color='green', label='Reviews')

ax.set(yticks=ind + width, yticklabels=top_rate['App'], ylim=[2*width - 1, len(top_rate)])
ax.legend()

plt.show()
```

For some reason, mutiple games are not consider as duplicated even after removal. However we can still see that most of them are games others are tools.
