```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("./Data/googleplaystore.csv")
data
```

For those who interested in the Mobile App data, We should probably dive deep into Each category and check the rating and number of reviews.

```python
cat_unique_count = data.Category.value_counts().rename_axis('Category').reset_index(name='counts')
cat_unique_count
```

```python
fig, ax = plt.subplots()
ax.pie(cat_unique_count.counts, labels=cat_unique_count.Category)
```

```python
genres_unique_count = data.Genres.value_counts().rename_axis('Genres').reset_index(name='counts')
genres_unique_count
count = genres_unique_count.counts
gen = labels=genres_unique_count.Genres

fig, ax = plt.subplots()
ax.pie(count, labels=gen)
```

In comparison we know that genres is more specific than Category, Lets pick Tools and Entertainment as target genre to analyze.

```python
data["Reviews"] = data["Reviews"].replace({"K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(int)
data["Installs"] = data["Installs"].str.replace(',', '')
data["Installs"] = data["Installs"].str.replace('+', '')
## remove only str value
data["Installs"] = data["Installs"].str.replace('Free', '')
pd.to_numeric(data["Installs"])
data
```

Lets first look under the tool genre

```python
tools_data = data[data["Genres"]=='Tools']
tools_data
```

Lets see how reviews and installs in groups are divided

```python
labels = ['no review','5000','10000','100000','1000000','10000000','100000000']
tools_data['grp_review'] = pd.cut(tools_data['Reviews'],
                   (0,5000,10000,100000,1000000,10000000,100000000,1000000000) , 
                   labels=labels)

plt.pie(tools_data.groupby(['grp_review'])['Reviews'].sum(), labels=labels)
```

For tools more nearly half of them has more than 100000000  reviews, most of them exceed 1000000 reviews

```python
ent_data = data[data["Genres"]=='Entertainment']
ent_data
```

```python
labels = ['no review','5000','10000','100000','1000000','10000000','100000000']
ent_data['grp_review'] = pd.cut(ent_data['Reviews'],
                   (0,5000,10000,100000,1000000,10000000,100000000,1000000000) , 
                   labels=labels)

plt.pie(ent_data.groupby(['grp_review'])['Reviews'].sum(), labels=labels)
```

For entertainment data most of them have more than 1000000 reviews and only very little amount of them have more than 10000000 reviews.
