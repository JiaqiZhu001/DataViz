import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

read dataset

```python
data = pd.read_csv("./Data/googleplaystore.csv")
data
```

As we can see, there are 13 columns including App name, Category, Reviews, Installation number.

THere 10840 observation in the dataset.

Lets take a look at how many Different Cataegory the dataset has.


```python
data.Category.unique()
```

```python
cat_unique_count = data.Category.value_counts().rename_axis('Category').reset_index(name='counts')
cat_unique_count
```

A pie chart might work on this situation if we want to check what portion each category takes.

```python
fig, ax = plt.subplots()
ax.pie(cat_unique_count.counts, labels=cat_unique_count.Category)
```

We can see most of the catgory, however, there are too many categories which makes those categories with smaller portion becomes unclear.

Switch to Histogram might be better.

```python
count = cat_unique_count.counts
cat = labels=cat_unique_count.Category
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(cat, count)
```

Horizontal Bar plot seems better. Also there is a very similar column called "Genres" We should also check that Column

```python
genres_unique_count = data.Genres.value_counts().rename_axis('Genres').reset_index(name='counts')
genres_unique_count
```

```python
count = genres_unique_count.counts
gen = labels=genres_unique_count.Genres
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(160, 90))
 
# Horizontal Bar Plot
ax.barh(gen, count)
```

Genres is more detailed than Categories.

Review is also a very important value for an App, usually more reviews means more installation

```python
plt.plot(data.App,data.Reviews)
plt.plot(data.App,data.Installs)
plt.show()
```

Plotting all observation turns out to be unreadable. Filter out some data will be a good option

```python
# data['Reviews'] = data['Reviews'].apply(value_to_float)
data["Reviews"] = data["Reviews"].replace({"K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(int)
data["Reviews"]
```

```python
reviews = data[data["Reviews"] > 10000000]['Reviews']
installs = data[data["Reviews"] > 10000000]['Installs']
apps = data[data["Reviews"] > 10000000]['App']
installs
```

```python
plt.plot(apps,reviews)
# plt.plot(apps,installs)
plt.show()
```

```python
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(apps, reviews)
```

Its better if we filter and use horizaontal bar graph.
